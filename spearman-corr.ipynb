```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from random import randint
```

```python
# data
x = range(0,101)
y_all = [
            [x for x in x], 
            [(x-50)**2 for x in x],
        ]
for y in y_all:
    # test spearman correalation
    corr = stats.spearmanr(x,y)
    plt.plot(x,y, label='Correlation = {:0.2f}'.format(corr.correlation))
plt.legend()
plt.show()
```
